```
import pyspark
```

```
df = spark.sql('SELECT * FROM customer_churn_csv')
```

```
df.printSchema()
```

```
df.show()
```

```
df.describe().show()
```

```
df.columns
```

```
my_data = df.select(['Age','Total_Purchase','Years', 'Num_Sites','Churn'])
```

```
from pyspark.ml.feature import (VectorIndexer, VectorAssembler)
```

```
assembler = VectorAssembler(inputCols=['Age','Total_Purchase','Years', 'Num_Sites'], outputCol='features')
```

```
assembler_output = assembler.transform(my_data)
```

```
assembler_output.show()
```

```
final_data = assembler_output.select('features','churn')
```

```
train,test = final_data.randomSplit([0.7,0.3])
```

```
from pyspark.ml.classification import LogisticRegression
```

```
log_model = LogisticRegression(featuresCol='features', labelCol='churn')
```

```
fitted_model = log_model.fit(train)
```

```
from pyspark.ml.evaluation import BinaryClassificationEvaluator # It checks the area under the curve (AUC)
```

```
results = fitted_model.evaluate(test)
```

```
results.predictions.show()
```

```
evaluator = BinaryClassificationEvaluator(rawPredictionCol='prediction', labelCol='churn')
```

```
auc = evaluator.evaluate(results.predictions)
```

```
auc
```

```
# Predictions on new data
new_data = spark.sql('SELECT * FROM new_customers_csv')
```

```
# Fitting the model on entire data 
final_model = log_model.fit(final_data)
```

```
new_data_assembler = assembler.transform(new_data)
```

```
new_data_assembler.show()
```

```
results = final_model.transform(new_data_assembler)
```
